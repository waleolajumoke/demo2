```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,9)
x
```

```python
y = x ** 2
y
```

```python
fig = plt.figure(figsize=(15,4))

# plotting a figure
plt.plot(x,y, marker="+", color="red")
plt.title("Graph Title")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.savefig("firstgraph.jpg")

# plt.show()
```

```python
plt.plot(x, x**2, label="2024 sales")
plt.plot(x, x**3, label="2023 sales")
plt.legend(loc="center left",bbox_to_anchor=(1, 0.5))
plt.text(1.1,0.6, 'Key', horizontalalignment='center', fontsize=12, transform=plt.gca().transAxes)
```

```python
df = pd.read_csv("student.csv")
df.head()
```

```python
df["total score"] = round(100 * (df["math score"] + df["reading score"] + df["writing score"])/ 300,2)
df.head()
```

```python
def grade(x):
    if x >= 90 and x <= 100:
        return "A"
    elif x >= 80 and x < 90:
        return "B"
    elif x >= 70 and x < 80:
        return "C"
    elif x >= 60 and x < 70:
        return "D"
    else:
        return "F"
```

```python
df["grade"] = df["total score"].apply(grade)
df.head()
```

```python
sns.countplot(df,x="grade")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(df, x="parental level of education", y="math score")
```
